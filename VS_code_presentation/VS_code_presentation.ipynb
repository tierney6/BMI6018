### What is VS Code?
Visual studio Code, which is typically referred to as VS Code is a type of code editor. It is developed by Microsoft, but is open-source, and has a wide variety of both built-in support or free extensions that users can download that support a large number of programming languages.

One of the main benefits is that it is a highly customizable to the user. Customizable features include color schemes, keyboard short cuts, and the UI interface.

It's UI layout can also incorporate the terminal, file explorer, and a remote explorer if you are running programs via SSH or on a server. 


### How to install VS Code
VS code is available for Mac, Linux, and Windows. It is free to use. 

VS Code is free to download and install. Installation instructions and downloads can be found [here](https://code.visualstudio.com/download).

### Creating a virtual environment

In order to run python code in VS code, it is good practice to create a virtual environment that is project-specific. This tells the executor which version of python that you would like to run your code with, along with any packages that you would like to use, and their versions.

This is especially a good practice for ensuring that your code is reproducible across computers, which is most important if you are working with collaborators. 

You can select an environment from the drop down menu at the top, or create a new one, using `venv` or `conda`.

The file extension will tell VS code how to interpret the code in each file, and any code executed within VS code will be run by the environment you selected above.
