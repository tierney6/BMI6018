```python

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Importing Data
flights_data = pd.read_csv('data/flights.csv')
flights_data.head(10)
weather_data_pd = pd.read_csv('data/weather.csv')
weather_data_np = weather_data_pd.to_numpy()
```

```python

# Pandas Data Filtering/Sorting Question Answering
# use flights_data

#Question 1 How many flights were there from JFK to SLC? Int

# first use .loc to subset the data to only include rows with JFK as origin and SLC as destination
jfk_slc = flights_data.loc[(flights_data['origin'] == "JFK") & (flights_data['dest'] == "SLC")]

# then since each row is a flight, get the number of rows to get the number of flights
# pandas' shape returns the number of rows and columns, we can extract the number of rows using [0]
q_1 = jfk_slc.shape[0]

print(f"There are {q_1} flights from JFK to SLC.")
```

```python

#Question 2 How many airlines fly to SLC? Should be int

# use .loc to again filter only rows with SLC as the destination airport
# then use ['origin'] to selct only the single Series that specifies the origin airport
origin_airports = flights_data.loc[(flights_data['dest'] == 'SLC')]['origin']


# now use .nunique to get the number of airports that fly to SLC
q_2 = origin_airports.nunique()

origin_airports.unique()
print(f"There are {q_2} airports that have flights into SLC in this data set: {origin_airports.unique()[0]}, and {origin_airports.unique()[1]}.")
```

```python

#Question 3 What is the average arrival delay for flights to RDU? float

# use .loc to subset flights arriving at RDU airport destination
# use ['arr_delay'] to select only the arrival delay column
# use .mean() to calculate the average values across this column
q_3 = flights_data.loc[flights_data['dest'] == "RDU"]['arr_delay'].mean()


print(f"The average delay of flights landing at RDU airport is {q_3:.2f} minutes")
```

```python
#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float

# use .loc to filter to flights with SEA as the destination airport
seattle = flights_data.loc[flights_data['dest'] == 'SEA']

# get a list of airports in NYC
nyc = ['JFK', 'LGA']
# filter the seattle dataframe to flights that originate in the nyc airports
seattle_nyc = seattle[seattle['origin'].isin(nyc)]
seattle_nyc

# get proportion of flights landing in seattle that originate from nyc airports

q_4 = seattle_nyc.shape[0]/seattle.shape[0]

f"The proportion of flights that land in Seattle that are from either JFK or LGA is {q_4:.2f}."
```

```python

#Question 5 Which date has the largest average depature delay? Pd slice with date and float
#please make date a column. Preferred format is 2013/1/1 (y/m/d)

# create a series called date 
# first convert the year, month, day values into strings from ints
date = flights_data[['year', 'month', 'day']].astype("str")
# then use .apply method to concanenate across each row (axis = 1) and join using "/" as a separator to get date = 'y/m/d'
date = date.apply('/'.join, axis=1)

# add date column back to original data set
flights_data['date'] = date

# get only delay column and date column
delay_data = flights_data[['date', 'dep_delay']]

# group by date, and get the mean dep_delay for each day
delay_mean = delay_data.groupby('date').mean()

# filter to the row in the data that is the maximum average delay
q_5 = delay_mean[delay_mean.dep_delay == delay_mean.dep_delay.max()]
q_5
print(f"The maximum delay was {q_5}")
```

```python

#Question 6 Which date has the largest average arrival delay? pd slice with date and float

# get only arrival delay column and date column
delayarr_data = flights_data[['date', 'arr_delay']]

# group by date, and get the mean arr_delay for each day
delayarr_mean = delayarr_data.groupby('date').mean()

# filter to the row in the data that is the maximum average delay
q_6 = delayarr_mean[delayarr_mean.arr_delay == delayarr_mean.arr_delay.max()]
q_6
print(f"The maximum arrival delay was {q_6}")
```

```python

#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
#speed = distance/airtime

# get speed by dividing distance by airtime
speed = flights_data['distance'] / flights_data['air_time'] 
# add speed as a column to the flight data
flights_data['speed'] = speed

# subset rows to only include where speed is at its max, and return only tailnum and speed columns
q_7 = flights_data[flights_data['speed'] == flights_data.speed.max()][['tailnum', 'speed']]

print(f"The fastest flight was {q_7}")
```

```python


#Question 8 Replace all nans in the weather pd dataframe with 0s. Pd with no nans
q_8 = weather_data_pd.fillna(0)
q_8
```

```python

# Numpy Data Filtering/Sorting Question Answering
#Use weather_data_np
#Question 9 How many observations were made in Feburary? Int
# month is the 4th column (3rd index)
# use [:,3] to return only the month column and all rows
# use == 2 to get a boolean only rows that have February as the month and sum since True = 1 and False = 0
q_9 = sum(weather_data_np[:,3] == 2)

print(f"Assuming each row is an observation, the number of observations made in February is {q_9}.")
```

```python
#Question 10 What was the mean for humidity in February? Float

# subset to get only february data
feb = weather_data_np[weather_data_np[:,3] == 2]

# get the 8th column for humidity and take the mean
q_10 = feb[:,8].mean()

print(f"The mean humidity for February is {q_10:.2f}.")
```

```python

#Question 11 What was the std for humidity in February? Float

# get the standard deviation for humidity in February
q_11 = feb[:,8].std()

print(f"The standard deviation for humidity for February is {q_11:.2f}.")
```
